```python
# import numpy
# from matplotlib import pyplot
# from mpl_toolkits.mplot3d import Axes3D
# from sklearn import decomposition, datasets

import numpy
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition,datasets
import pandas as pd
```

```python
#3d and 3 component
df_data = pd.read_csv("titanic.csv")
df = df_data[["Survived","Pclass","Age","Fare","Sex"]].dropna()
del df["Sex"]
x  = df.values

y = df_data[["Survived","Pclass","Age","Fare","Sex"]].dropna()
y["label"] = 1
y.loc[y['Sex'] == 'female', 'label'] = 0
y = y["label"].values

figure = pyplot.figure(1,figsize=(4,3))
pyplot.clf()
axes = Axes3D(figure,rect=[0,0,.95,1],elev=48, azim=134)
axes.cla()

pca = decomposition.PCA(n_components=3)
pca.fit(x)
x = pca.transform(x)

axes.scatter(x[:,0],x[:,1],x[:,2],c=y,cmap=pyplot.cm.nipy_spectral, edgecolor='k')
pyplot.show()
```

```python
#2d and 2 component

df_data = pd.read_csv("titanic.csv")
# df_data.head()
df = df_data[["Survived","Pclass","Age","Fare","Sex"]].dropna()
del df["Sex"]
x  = df.values

y = df_data[["Survived","Pclass","Age","Fare","Sex"]].dropna()
y["label"] = 1
y.loc[y['Sex'] == 'female', 'label'] = 0
y = y["label"].values


pca = decomposition.PCA(n_components=2)
pca.fit(x)
x = pca.transform(x)

pyplot.figure(figsize=(8,6))    
pyplot.scatter(x[:,0], x[:,1],marker='o', 
            color='blue', alpha=0.7, label='x1 samples')

pyplot.title('Basic scatter plot')
pyplot.ylabel('variable X')
pyplot.xlabel('Variable Y')
# pyplot.legend(loc='upper right')
pyplot.show()
```
